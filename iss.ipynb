```python
import requests
import matplotlib as plt
from ast import literal_eval
import time
datapoints = {}
times = []
```

```python
def points():
    for i in range(0,150):
        rep = requests.get("http://api.open-notify.org/iss-now.json")
        point = rep.content
        datapoints[i] = (literal_eval(point.decode()))
        time.sleep(10)
    print(datapoints)
        
```

```python
import datetime

lats = []
longs = []
def latlong():
    for i in range(0,len(datapoints)):
        lats.append(datapoints[i]['iss_position']['latitude'])
        longs.append(datapoints[i]['iss_position']['longitude'])
    
```

```python
timee = []
def timeelapsed():
    for i in range(0, len(datapoints)):
        unixv = (datapoints[i]['timestamp'])-1688828935
        timee.append(unixv)
    print(timee)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.subplot(1,2,1)
plt.plot(timee,lats)
plt.yticks(lats[::5])
plt.xlabel("Time Elapsed")
plt.ylabel("Latitude")

plt.subplot(1,2,2)
plt.plot(timee,longs)
plt.yticks(longs[::5])
plt.xlabel("Time Elapsed")
plt.ylabel("Longitude")
plt.show
```
